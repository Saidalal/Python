```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 ***Replacing blanks with 0***

```python
df['TotalCharges'] = df["TotalCharges"].replace(" ",0).astype("float")
```

***Convert 0 and 1 values of seniorCitizen toyes/no to make it easier to understand***

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

***TOP 5***

```python
df.head()
```

***Number of customers who churned vs. those who stayed***

```python
ax = sns.countplot(x='Churn', data=df)

ax.bar_label(ax.containers[0])
plt.title("Count of customers by Churn")
plt.show()
```

***From the given pie-chart we can conclude 26.54% of customers have churned***

```python
plt.figure(figsize=(4,4))
gb = df.groupby("Churn").agg({'Churn': 'count'})
plt.pie(gb['Churn'],labels = gb.index ,autopct = "%1.2f%%")
plt.title("Percentage of Churn")
plt.show()
```

***customer churn counts by gender***

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(x='gender', hue='Churn', data=df, palette=['#E1AD01', '#088F8F'])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Count of customers by gender and churn")
plt.show()


```

***Plot stacked bar chart showing churn percentages by senior citizen status with labels.***


```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

ax = total_counts.plot(kind='bar', stacked=True, figsize=(5, 4), color=['#1f77b4', '#ff7f0e'])

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_y() + p.get_height()/2, f'{p.get_height():.1f}%', ha='center', va='center')

ax.set(title='Churn by Senior Citizen (Stacked Bar Chart)', xlabel='SeniorCitizen', ylabel='Percentage (%)')
ax.legend(title='Churn', bbox_to_anchor=(0.9, 0.9))
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

***Plot the distribution of customer tenure (in months) separated by churn status***


```python
plt.figure(figsize=(9,4))
sns.histplot(data=df, x = 'tenure', hue='Churn', bins=72)
plt.title('Distribution of Tenure by Churn')
plt.xlabel('Tenure (Months)')
plt.ylabel('Count')
plt.show()
```

***People  who have month-to-month contract are likely to churn then from those who have 1 or 2 years of contract.***

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(x = 'Contract' ,data = df,hue = 'Churn')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])

plt.title("Count of customers by Contract")
plt.show()
```

***Boxplot to compare the distribution of monthly charges between churned and non-churned customers***

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette=['#E1AD01', '#088F8F'])
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.title('Monthly Charges vs Churn')
plt.show()
```

 ***Subplots to visualize countplots for multiple categorical features,
 showing their distributions by churn status***

```python

service_cols = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
                'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']


fig, axes = plt.subplots(3, 3, figsize=(18, 12))  # 3 rows x 3 columns

for ax, col in zip(axes.flatten(), service_cols):
    sns.countplot(x=col, hue='Churn', data=df, ax=ax, palette=['#E1AD01', '#088F8F'])
    ax.set_title(f'{col} vs Churn')


plt.tight_layout()
plt.show()
```

***Donut chart showing the proportion of churned vs non-churned customers***


```python
import matplotlib.pyplot as plt

# Count churn values
churn_counts = df['Churn'].value_counts()

# Create pie chart
plt.figure(figsize=(5,5))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.2f%%', startangle=90, colors=['#E1AD01', '#088F8F'])


centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

plt.title('Customer Churn Distribution (Donut Chart)')
plt.tight_layout()
plt.show()
```
